```python
import shutil
from pathlib import Path

brisc_seg = Path("/kaggle/input/datasets/briscdataset/brisc2025/brisc2025/segmentation_task")
OUT_ROOT = Path("/kaggle/working/segmentation")
out_img = OUT_ROOT / "brisc" / "images"
out_msk = OUT_ROOT / "brisc" / "masks"
out_img.mkdir(parents=True, exist_ok=True)
out_msk.mkdir(parents=True, exist_ok=True)

count, skipped = 0, 0
for split in ["train", "test"]:
    img_dir = brisc_seg / split / "images"
    msk_dir = brisc_seg / split / "masks"
    for f in sorted(img_dir.glob("*")):
        mask_file = msk_dir / (f.stem + ".png")
        if (out_img / f.name).exists():
            skipped += 1
            continue
        if mask_file.exists():
            shutil.copy2(f, out_img / f.name)
            shutil.copy2(mask_file, out_msk / mask_file.name)
            count += 1

print(f"✅ {count} pairs copied")
print(f"⏭️  {skipped} skipped (trùng tên)")
print(f"   images: {len(list(out_img.glob('*')))}, masks: {len(list(out_msk.glob('*')))}")

# Xem mẫu
for sub in ["images", "masks"]:
    files = sorted((OUT_ROOT / "brisc" / sub).glob("*"))
    print(f"\n📁 brisc/{sub}/ ({len(files)} files)")
    for f in files[:3]:
        print(f"  {f.name}")
```

```python
import shutil
from pathlib import Path

NB_ROOT = Path("/kaggle/input/notebooks/ruby292/eda-5-datasets/segmentation")
OUT_IMG = Path("/kaggle/working/merged/images")
OUT_MSK = Path("/kaggle/working/merged/masks")
OUT_IMG.mkdir(parents=True, exist_ok=True)
OUT_MSK.mkdir(parents=True, exist_ok=True)

# Định nghĩa nguồn: (img_dir, msk_dir)
sources = [
    (NB_ROOT / "Glioma" / "images",         NB_ROOT / "Glioma" / "masks"),
    (NB_ROOT / "Meningioma" / "images",      NB_ROOT / "Meningioma" / "masks"),
    (NB_ROOT / "Pituitary tumor" / "images", NB_ROOT / "Pituitary tumor" / "masks"),
    (NB_ROOT / "anhxunv" / "images",         NB_ROOT / "anhxunv" / "masks"),
    (NB_ROOT / "nikhil_images",              NB_ROOT / "nikhil_masks"),
    (Path("/kaggle/working/segmentation/brisc/images"), Path("/kaggle/working/segmentation/brisc/masks")),
]

total, skipped = 0, 0
for img_dir, msk_dir in sources:
    label = img_dir.parent.name if img_dir.name == "images" else img_dir.name
    count = 0
    for f in sorted(img_dir.glob("*")):
        if (OUT_IMG / f.name).exists():
            skipped += 1
            continue
        # Tìm mask: cùng tên hoặc đổi đuôi .png
        msk_file = msk_dir / f.name
        if not msk_file.exists():
            msk_file = msk_dir / (f.stem + ".png")
        if not msk_file.exists():
            msk_file = msk_dir / (f.stem + "_mask.png")
        if msk_file.exists():
            shutil.copy2(f, OUT_IMG / f.name)
            shutil.copy2(msk_file, OUT_MSK / msk_file.name)
            count += 1

    print(f"✅ {label}: {count} pairs")
    total += count

print(f"\n📊 Tổng: {total} pairs, {skipped} skipped (trùng tên)")
print(f"   images: {len(list(OUT_IMG.glob('*')))}")
print(f"   masks:  {len(list(OUT_MSK.glob('*')))}")
```

```python
import json, os
from pathlib import Path

DATASET_DIR = Path("/kaggle/working/merged")

# Verify trước khi upload
imgs = len(list((DATASET_DIR / "images").glob("*")))
msks = len(list((DATASET_DIR / "masks").glob("*")))
print(f"images: {imgs}, masks: {msks}")

if imgs > 0 and msks > 0:
    meta = {
        "title": "brain-tumor-segmentation-merged",
        "id": "ruby292/brain-tumor-segmentation-merged",
        "licenses": [{"name": "CC0-1.0"}]
    }
    with open(DATASET_DIR / "dataset-metadata.json", "w") as f:
        json.dump(meta, f)

    os.system(f"kaggle datasets version -p {DATASET_DIR} -m 'full re-upload: images + masks + manifest' --dir-mode zip")
    print("✅ Uploaded!")
else:
    print("❌ DỪNG — thiếu file, không upload")
```
